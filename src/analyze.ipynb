```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/wittich/Library/Mobile Documents/com~apple~CloudDocs/Downloads/muon_data_20241004_0923.csv', sep=', +', engine='python')
```

```python
df.head()
```

```python
df['t'] = pd.to_datetime(df['t'], unit='ms')
df.set_index('t', inplace=True)
```

```python
df['dt'].plot.hist(bins=100, alpha=0.5)
```

```python
df['V'].plot.hist(bins=100)
```

```python
df[df['dt'] > 20]['V'].plot.hist(bins=100)
plt.xlabel('V')
plt.ylabel('Frequency')
plt.title(f'Histogram of V for dt > 20 (n={len(df[df["dt"] > 20])})')
plt.show()
```
